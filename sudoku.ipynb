## Part One
You should write all of your code for solving sudokus below this cell.

You must include a function called `sudoku_solver(sudoku)` which takes one sudoku puzzle (a 9x9 NumPy array) as input, and returns the solved sudoku as another 9x9 NumPy array. This is the function which will be tested. 

```python
import numpy as np

def find_empty_location(board):
    """"
    Find the first empty cell in the board and return its row and column indices
    return tuple of row and column indices of the first empty cell
    in the board,
    otherwise returns None if the board is full.
    """
    for i in range(9):
        for j in range(9):
            if board[i][j] == 0:
                return i, j
    return None

def is_valid(board, row, col, num):
    """
    Checks if it is safe to place a number in a specific cell and
    returns true if safe to do so else returns false.
    """
    # Checks if number is in the same row
    if num in board[row]:
        return False
    # Checks if number is in the same column
    for i in range(9):
        if board[i][col] == num:
            return False
    # Checks if num is in the same 3x3 box
    box_row = row // 3
    box_col = col // 3
    for i in range(3 * box_row, 3 * box_row + 3):
        for j in range(3 * box_col, 3 * box_col + 3):
            if board[i][j] == num:
                return False
    # All checks passes, so number can be placed in the cell
    return True

def apply_constraints(board):
    """
    Applies constraints to the board by iterating 
    through the cells and checking if a given cell has 
    only one possible value based on the current state 
    of the board. If it does, then the value of the 
    cell is updated.
    """
    changed = True
    # Keeps applying constraints until no more changes can be done
    while changed:
        changed = False
        # Intialize updated_board with a copy of board
        updated_board = board.copy()
        for i in range(9):
            for j in range(9):
                # Skips over cells that already have values
                if board[i][j] != 0:
                    continue
                # Determine the possible values for the cell based on the current state of the board
                possible_values = set(range(1,10))
                for k in range(9):
                    if board[i][k] in possible_values:
                        possible_values.remove(board[i][k])
                    if board[k][j] in possible_values:
                        possible_values.remove(board[k][j])
                box_row = i // 3
                box_col = j // 3
                for m in range(3 * box_row, 3 * box_row + 3):
                    for n in range(3 * box_col, 3 * box_col + 3):
                        if board[m][n] in possible_values:
                            possible_values.remove(board[m][n])
                # If the cell has only one possible value, set the cell's value and update board
                if len(possible_values) == 0:
                    return np.full((9,9), -1)
                elif len(possible_values) == 1:
                    board[i][j] = list(possible_values)[0]
                    changed = True       
    # Returns updated_board after making changes to it
    return updated_board

def solved_sudoku(board):
    """ 
    Solves the soduku board using backtracking.
    board is a list of lists of digits, where 0 represents an empty cell
    returns the solved sudoku 
    if there are no solutions it returns the sudoku filled with -1s.
    """
    # Apply constraints until no more changes can be made to the board
    while True:
        updated_board = apply_constraints(board)
        if np.array_equal(updated_board, board):
            break
        board[:] = updated_board[:]

    location = find_empty_location(board)
    if not location:
        # Board is full, solution found
        return board
    
    # Use backtracking to search for solution
    row, col = location
    for num in range(1, 10):
        if is_valid(board, row, col, num):
            board[row][col] = num
            solution = solved_sudoku(board)
            if solution is not None:
                return solution
            # Reset the cell to 0 and try the next number
            board[row][col] = 0    
    # No solution found, returns a sudoku filled with -1s
    return np.full((9,9), -1)

def check_board(board):
    """
    Checks final board and returns it if it is a valid solution
    otherwise return the sudoku filled with -1s.
    """
    # Checks if final board has any empty cells
    if np.any(board == 0):
        return np.full((9,9), -1)
    # Checks if any rows have duplicate values         
    for row in board:             
        if len(set(row)) < 9:                 
            return np.full((9,9), -1)         
    # Checks if any columns have dupicate values         
    for col in range(9):             
        if len(set(board[:,col])) < 9:                 
            return np.full((9,9), -1)
    # Checks if any 3x3 has duplicate values
    for i in range(3):
        for j in range(3):
            box = board[i * 3:(i + 1) * 3, j * 3:(j + 1) * 3]
            if len(set(box.flatten())) < 9:
                return np.full((9,9), -1)
    return board
    
def sudoku_solver(sudoku):
    """
    Main function that outputs the solved sudoku or if there
    is no solution a sudoku filled with -1s.
    """
    return check_board(solved_sudoku(sudoku))
```
